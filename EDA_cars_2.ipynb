# Câu 3. 
Tính correlation của các feature = ['bore', 'stroke', 'compression-ratio','horsepower']. Nhận xét xem có mối quan hệ tương quan nào không?

```python
data_cor = data[['bore', 'stroke', 'compression-ratio','horsepower']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corrmat = data_cor.corr()
f, ax = plt.subplots(figsize =(9, 8))
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1, annot=True)
```

## Nhận xét 
- Xét những giá trị có giá trị lớn hơn 0.3 sẽ có tương quan
=> Những feature có mối tương quan là: horsepower và bore

# Câu 5. 
Xét toàn bộ dataset, tìm các feature (kiểu số) có correlation so với price có ảnh hưởng nhiều nhất (đồng thuận, hoặc không đồng thuận)

```python
feature_num = []
for i in range(0, len(data.dtypes.values)):
    if data.dtypes.values[i] == "int64" or data.dtypes.values[i] == "float64":
        feature_num.append(data.dtypes.index[i])
```

```python
data_cor_5 = data[feature_num]
```

```python
corrmat_5 = data_cor_5.corr()
```

```python
corrmat_5
```

```python
f5, ax5 = plt.subplots(figsize =(20, 20))
sns.heatmap(corrmat_5, ax = ax5, cmap ="YlGnBu", linewidths = 0.1, annot=True)
```
